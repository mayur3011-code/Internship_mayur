# Q2)Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)
# and make data frame.


```python
# Function Definition
def IMDB_top_100(url):
    r=requests.get(url)
    soup=BeautifulSoup(r.content,'html.parser')
    name=[]
    year=[]
    rating=[]
    for i in soup.find_all('img'):
        name.append(i.attrs['alt'])
    for i in list(soup.find_all('span',attrs={"secondaryInfo"})):
        year.append(i.text)
    for i in soup.find_all('strong'):
        rating.append(float(i.text))
    df=pd.DataFrame({'names':name[:100],
                     'years':year[:100],
                     'ratings':rating[:100]})
    print(df)
    df.to_csv('IMDB_top_100.csv', index = False)
```

```python
# Calling Function
IMDB_top_100("https://www.imdb.com/chart/top/")
```

# Q4) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
# a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.


```python
# Defining the function
def Top_10_ODI_team(url):
    r=requests.get(url)
    soup=BeautifulSoup(r.content,'html.parser')
    team=[]
    match=[]
    rating=[]
    for i in soup.find_all('span',attrs={'u-hide-phablet'}):
        team.append(i.text)
    match.append(soup.find('td',attrs={'rankings-block__banner--matches'}).text)
    rating.append(soup.find('td',attrs={'rankings-block__banner--points'}).text)
    count=0
    for i in list(soup.find_all('td',attrs={'table-body__cell u-center-text'}))[:18]:
        if count%2==0:
            match.append(i.text)
        else:
            rating.append(i.text)
        count+=1
    df=pd.DataFrame({'Team':team[:10],
                     'Match':match,
                     'Points':rating})
    df.to_csv('top_10_odi_teams_men.csv', index = False)
    print(df)
```

```python
Top_10_ODI_team('https://www.icc-cricket.com/rankings/mens/team-rankings/test')
```

# b ) Top 10 ODI Batsmen in men along with the records of their team and rating.

```python
def Top_10_ODI_batsman(url):
    r=requests.get(url)
    soup=BeautifulSoup(r.content,'html.parser')
    name=[]
    team=[]
    point=[]
    name.append(soup.find('div',attrs={"rankings-block__banner--name-large"}).text)
    for i in list(soup.find_all('td',attrs={'table-body__cell rankings-table__name name'}))[:9]:
        name.append(i.text.replace('\n',''))
    team.append(soup.find('div',attrs={"rankings-block__banner--nationality"}).text)
    for i in list(soup.find_all('td',attrs={'table-body__cell nationality-logo rankings-table__team'}))[:9]:
        team.append(i.text.replace('\n',''))
    point.append(soup.find('div',attrs={"rankings-block__banner--rating"}).text)
    for i in list(soup.find_all('td',attrs={'table-body__cell rating'}))[:9]:
        point.append(i.text)
    df=pd.DataFrame({'Name': name,
                     'Team':team,
                     'points':point})
    df.to_csv('top_10_odi_batsman.csv', index = False)
    print(df)
```

```python
Top_10_ODI_batsman("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting")
```

# 8) Write a python program to scrape mentioned details from dineout.co.in :
# # i) Restaurant name
## ii) Cuisine
## iii) Location
## iv) Ratings
## v) Image URL

```python
# i) To extract list of restaurants
def Restaurant_name(url):
    page=requests.get(url)
    source=page.content
    soup=BeautifulSoup(source)
    restaurant=[]
    for i in soup.find_all('a','restnt-name ellipsis'):
        restaurant.append(i.text)
    return pd.DataFrame({'Restaurant name':restaurant}) 

   
```

```python
Restaurant_name('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

```python
# ii) To extract cuisine
def cuisine_details(url):
    page=requests.get(url)
    source=page.content
    soup=BeautifulSoup(source)
    cuisine=[]
    for i in list(soup.find_all('span',attrs={'double-line-ellipsis'})):
        cuisine.append(i.text.split('|')[1])
        
    return pd.DataFrame({'Cuisine details':cuisine})
```

```python
cuisine_details('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

```python
# # iii) To extract locations
def locations (url):
    page=requests.get(url)
    source=page.content
    soup=BeautifulSoup(source)
    locations = []
    for i in soup.find_all('div','restnt-loc ellipsis'):
            locations.append(i.text)
    return pd.DataFrame({'Locations':locations}) 
```

```python
locations('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

```python
# iv) To extract Ratings
def Ratings (url):
    page=requests.get(url)
    source=page.content
    soup=BeautifulSoup(source)
    ratings =[]
    for i in soup.find_all('div','restnt-rating rating-4'):
        ratings.append(i.text)
    return pd.DataFrame({'Ratings':ratings}) 
```

```python
Ratings('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

```python
# v) To extract Image Url
def Image_url (url):
    page=requests.get(url)
    source=page.content
    soup=BeautifulSoup(source)
    Imageurl =[]
    for i in soup.find_all('img','no-img'):
        Imageurl.append(i.text)
    return pd.DataFrame({'Image URL':Imageurl})
```

# 10) Write a python program to scrape monument name, monument description, image URL about top 10 monuments
# from puredestinations.co.uk

```python
import re
def  puredestinations(url):
    page=requests.get(url)
    source=page.content
    soup=BeautifulSoup(source,'html.parser')
    
    name=[]
    desc=[]
    url=[]
    for i in list(soup.find_all('strong')):
        name.append(i.text)
    for i in list(soup.find_all('p')):
        desc.append(i.text)
    for i in list(soup.find_all('img',attrs={'srcset': re.compile("^https://")})):
        url.append(i.get('srcset'))
    return pd.DataFrame({'Monument Name':name[:10],'Description':desc[5:33:3],'Url':url[:10]})
        
```

```python
puredestinations('https://www.puredestinations.co.uk/top-10-famous-monuments-to-visit-in-india/')
```
